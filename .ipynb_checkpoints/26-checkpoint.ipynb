```python
# Перше завдання

import math

# Імовірність того, що телевізор витримає гарантійний термін роботи
p = 0.9

# Кількість телевізорів
n = 100

# Мінімальна кількість телевізорів, які витримають гарантійний термін роботи
min_tv = 85

# Максимальна кількість телевізорів, які витримають гарантійний термін роботи
max_tv = 95

# Обчислюємо ймовірність того, що кількість телевізорів, які витримають гарантійний термін роботи, буде в межах [min_tv; max_tv]
probability = sum([math.comb(n, k) * (p ** k) * ((1-p) ** (n-k)) for k in range(min_tv, max_tv+1)])

print(f"Ймовірність того, що кількість телевізорів, які витримають гарантійний термін роботи, буде в межах [85;95] дорівнює {probability:.2%}")

```

```python
# Друге завдання

import numpy as np
import matplotlib.pyplot as plt

# Вибірка
data = [10, 13, 10, 9, 9, 12, 12, 6, 7, 9, 8, 9, 11, 9, 14, 13, 9, 8, 8, 7, 10, 10, 11, 11, 11, 12, 8, 7, 9, 10, 14, 13, 8, 8, 9, 10, 11, 11, 12, 12]

# Розрахунок дискретного статистичного розподілу
unique_values, counts = np.unique(data, return_counts=True)
probabilities = counts / len(data)

# Вивід дискретного статистичного розподілу
print("x\tP(X=x)")
for i in range(len(unique_values)):
    print("{}\t{}".format(unique_values[i], probabilities[i]))

# Побудова полігону частот
plt.hist(data, bins=range(min(data), max(data)+2), density=True, alpha=0.5, color='blue', edgecolor='black')
plt.ylabel('Частота')
plt.xlabel('Значення')
plt.title('Полігон частот')
plt.show()

# Побудова F*(x)
cumulative_probabilities = np.cumsum(probabilities)
plt.step(unique_values, cumulative_probabilities, where='post', color='blue', alpha=0.5)
plt.ylabel('F*(x)')
plt.xlabel('Значення')
plt.title('Емпірична функція розподілу')
plt.show()

# Обчислення моди та медіани
mode = unique_values[np.argmax(counts)]
median = np.median(data)
print("Мода: {}".format(mode))
print("Медіана: {}".format(median))

```
